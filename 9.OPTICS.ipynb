```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import OPTICS
optics=OPTICS(metric='minkowski',algorithm='brute',min_samples=15,eps=5)
y_means=optics.fit_predict(x)
```

```python
y_means
```

```python
supervised=(dataset)
```

```python
supervised['cluster_group']=y_means
```

```python
supervised
```

```python
centroids=optics.labels_
```

```python
centroids
```

```python
cluster_centers_indices=optics.reachability_
print(cluster_centers_indices)
```

```python
n_clusters_=len(cluster_centers_indices)
```

```python
print(n_clusters_)
```

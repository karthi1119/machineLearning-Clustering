```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(metric='euclidean',algorithm='ball_tree',min_samples=5)
```

```python
y_means=dbscan.fit_predict(x)
```

```python
y_means
```

```python
supervised=(dataset)
```

```python
supervised['cluster_group']=y_means
```

```python
supervised
```

```python
n_c=dbscan.components_
n_c
```

```python
centroids=dbscan.labels_
```

```python
centroids
```

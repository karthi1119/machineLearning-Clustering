```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
#!pip install HDBSCAN
```

```python
import hdbscan as hd
# Assuming x is your data
clusterer = hd.HDBSCAN()
y_means = clusterer.fit_predict(x)
```

```python
y_means
```

```python
supervised=(dataset)
```

```python
supervised['cluster_group']=y_means
```

```python
supervised
```

```python
centroids=clusterer.labels_
```

```python
centroids
```

```python
cluster_centers_indices=clusterer.probabilities_
print(cluster_centers_indices)
```

```python
n_clusters_=len(cluster_centers_indices)
```

```python
print(n_clusters_)
```
